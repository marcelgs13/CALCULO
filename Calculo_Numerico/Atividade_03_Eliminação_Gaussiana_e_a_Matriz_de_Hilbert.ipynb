```python
def eliminacao_gaussiana(A,b):
  n = len(b)

  for k in range(0, n-1):
    pivo = A[k][k]
#percorre as linhas abaixo do pivo
    for i in range(k+1, n):
      #multiplicador
      m = A[i][k]/pivo

      #percorre as colunas
      for j in range(k, n):
        A[i][j] = A[i][j] - m*A[k][j]

      b[i] = b[i] - m*b[k]

  det = 1
  for  t in range(n):
    det = det * A[t][t]




  return A, b, det # depois de escalonar
```

```python
def subs_reversa(A_t, b_t):
  n = len(b_t)

  x = n*[0]

  x[n-1] = b_t[n-1] / A_t[n-1][n-1]
#percorre as linhas acima at√© a primeira
  for k in range(n-2, -1, -1):
    s= 0
    for j in range(k+1, n):
      s += A_t[k][j] * x[j]

    x[k] = (b_t[k] - s)/ A_t[k][k]

  return x
```
